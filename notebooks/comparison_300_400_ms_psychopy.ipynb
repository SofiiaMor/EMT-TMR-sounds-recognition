```python
# all imports
import io
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# upload all files
from google.colab import files
uploaded = files.upload()
```

```python
# output table with all subjects
output = []
columns_name = ('subject', 'mean_accuracy_300', 'mean_accuracy_400', 'mean_RT_300', 'mean_RT_400')

# go through all xls files
for i in uploaded:
  # read xls file
  df = pd.read_excel(io.BytesIO(uploaded.get(i)))
  subject_id = i.split('_')[2].split('.')[0]
  # mean accuracy for 300 and 400ms sounds
  accur300 = df[df['condition'] == 300]['key_resp.corr'].mean()
  accur400 = df[df['condition'] == 400]['key_resp.corr'].mean()
  # mean RT for 300 and 400 ms sounds
  rt300 = df[df['condition'] == 300]['key_resp.rt'].mean()
  rt400 = df[df['condition'] == 400]['key_resp.rt'].mean()

  # add all data in row
  output.append((subject_id, accur300, accur400, rt300, rt400))

# create dataframe from all data
final_df = pd.DataFrame(output, columns=columns_name)
final_df
```

```python
# t test to compare accuracy for 300 and 400 ms
accur_stat = stats.ttest_rel(final_df['mean_accuracy_300'], final_df['mean_accuracy_400'])
```

```python
# t test to compare RT for 300 and 400 ms
rt_stat = stats.ttest_rel(final_df['mean_RT_300'], final_df['mean_RT_400'])
```

```python
final_df.at[0, 't_statistic_accur'] = accur_stat[0]
final_df.at[0,'p-value_accur'] = accur_stat[1]
final_df.at[0,'t_statistic_rt'] = rt_stat[0]
final_df.at[0,'p-value_rt'] = rt_stat[1]
```

```python
# export to xls
file_name = 'psychopy 300 and 400ms t-statistic.xlsx'
final_df.to_excel(file_name)

print('DataFrame is written to Excel File successfully.')

# download to the local computer
files.download(file_name)
```
